## 테이블에 대한 연산

### GROUP BY

split -> apply -> combine 과정을 거쳐 연산
- split : 한 시리즈에서 같은 인덱스 값을 가지는 값끼리 시리즈의 형태로 모음
- apply : 각각의 시리즈에 함수를 적용시킴
- combine : 연산된 결과값을 하나의 시리즈로 다시 합침

#### **group by basic**

```python
# data from: 
ipl_data = {'Team': ['Riders', 'Riders', 'Devils', 'Devils', 'Kings',
        'Kings', 'Kings', 'Kings', 'Riders', 'Royals', 'Royals', 'Riders'],
        'Rank': [1, 2, 2, 3, 3,4 ,1 ,1,2 , 4,1,2],
        'Year': [2014,2015,2014,2015,2014,2015,2016,2017,2016,2014,2015,2017],
        'Points':[876,789,863,673,741,812,756,788,694,701,804,690]}

df = pd.DataFrame(ipl_data)
df
```

```python
df.groupby("Team")["Points"].sum() # dataframe.groupby(묶음의 기준이 되는 열)[연산을 적용할 열].적용할 연산()
```

**hierarchical index** : group by의 결과로 두 개 이상의 인덱스가 생성됨

```python
h_index = df.groupby(["Team", "Year"])["Points"].sum() # 묶음의 기준이 되는 열을 여러개 지정 가능
h_index
```

```python
h_index.index
```

```python
h_index.unstack() # 그룹으로 묶여진 데이터를 행렬의 형태로 전환
```

```python
h_index.reset_index() # 인덱스를 없앨 수도 있음
```

```python
h_index.swaplevel() # 인덱스간의 레벨을 바꿀 수 있음
```

**group by 결과물의 정렬**

```python
h_index.swaplevel().sort_index(level=0) # 인덱스를 기준으로 정렬
```

**transfromation**
</br> 요약된 정보가 아니라 개별의 값에 지정해줌

```python
score = lambda x: (x)
grouped.transform(score)
```

단, max나 min처럼 시리즈 데이터에 적용되는 데이터들은 그룹으로 묶인 데이터를 기준으로 출력함

```python
grouped["Points"].max().sort_values()
```

```python
score = lambda x: (x.max())
grouped.transform(score).sort_values("Points") # 같은 팀일 경우에는 포인트가 동일한 값으로 출력됨
```

(ex) 정규화

```python
score = lambda x: (x - x.mean()) / x.std()
grouped.transform(score)
```

(ex) 최댓값을 1, 최솟값을 0으로 변환

```python
score = lambda x: (x - x.min()) / (x.max() - x.min())
grouped.transform(score)
```

### Pivot table & Crosstab

#### **Pivot table**


- 인덱스 축은 Group by와 동일함
- 각 열에 라벨링 값을 추가하여 값에 numeric type 값을 aggregation 하는 형태

```python
df_phone.head()
```

```python
df_phone.pivot_table(["duration"], # 연산을 적용시킬 열
                    index=[df_phone.month,df_phone.item], columns=df_phone.network, # 그룹으로 묶을 기준이 되는 열들
                    aggfunc="sum", fill_value=0) # 적용할 연산, 연산이 되지 않을 경우 0값을 채워넣음
```

#### **crosstab**

```python
df_movie = pd.read_csv("./data/movie_rating.csv")
df_movie.head()
```

```python
df_movie.pivot_table(["rating"],
                    index=df_movie.critic,
                    columns=df_movie.title, aggfunc="sum", fill_value=0)
```

```python
pd.crosstab(index=df_movie.critic, columns=df_movie.title, # 그룹으로 묶을 기준이 되는 열들
            values=df_movie.rating, aggfunc="first").fillna(0) # 연산을 적용시킬 열과 적용할 연산, 연산이 안 되는 경우 0을 채워넣음
```

```python
df_movie.groupby(["critic","title"]).agg({"rating":"first"}).unstack().fillna(0)
```
